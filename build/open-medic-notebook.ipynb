```python
### Ameli Open Medic: filter ambulatory drug prescriptions by physician specialty

import pandas as pd

# 1 & 2. Open the CSV file with UTF-8 encoding, ";" separator, and default first-line headers
file_path = "OPEN_MEDIC_2024.csv"
df = pd.read_csv(file_path, sep=";", encoding="latin-1", decimal=",", thousands=".")

# 3. Only keep the specified columns
columns_to_keep = ["CIP13", "l_cip13", "age", "sexe", "PSP_SPE", "BOITES", "REM"]
df = df[columns_to_keep]

# 4. Filter the rows where PSP_SPE is 15 (Ophtalmologists)
# Note: If PSP_SPE is read as text in your dataset, change 15 to "15"
df_filtered = df[df["PSP_SPE"] == 15]

# 5. Merge rows sharing the same CIP13 value and sum BOITES and REM
# We include 'l_cip13' (the text label) in the groupby so you don't lose the drug names.
# Age and sexe are automatically dropped here as they vary within the same CIP13 group.
df_final = (
    df_filtered.groupby(["CIP13", "l_cip13"])
    .agg({"BOITES": "sum", "REM": "sum"})
    .reset_index()
    .sort_values(by="BOITES", ascending=False)
)
df_final["REM"] = df_final["REM"].round(0).astype(int)
df_final = df_final.reset_index(drop=True)

# Save
output_file = "OPEN_MEDIC_2024_OPHT_FILTERED.csv"
df_final.to_csv(output_file, sep=";", index=False, encoding="utf-8")
print(f"Success! Processed data exported to: {output_file}")
df_final.head()
```
